```python
import os
import random
import shutil
from math import floor

# Specify full paths for your directories
shuffled_images_path = '/home/trainee/Desktop/Final_Work/yolov5-master-shuffled(after-aug)/YOLO-DATASET/images/shuffled_images'
shuffled_labels_path = '/home/trainee/Desktop/Final_Work/yolov5-master-shuffled(after-aug)/YOLO-DATASET/labels/shuffled_labels'
train_images_path = '/home/trainee/Desktop/Final_Work/yolov5-master-shuffled(after-aug)/YOLO-DATASET/images/train'
val_images_path = '/home/trainee/Desktop/Final_Work/yolov5-master-shuffled(after-aug)/YOLO-DATASET/images/val'
train_labels_path = '/home/trainee/Desktop/Final_Work/yolov5-master-shuffled(after-aug)/YOLO-DATASET/labels/train'
val_labels_path = '/home/trainee/Desktop/Final_Work/yolov5-master-shuffled(after-aug)/YOLO-DATASET/labels/val'

# Set a seed for reproducibility
random_seed = 42
random.seed(random_seed)

# Check if directories exist
print(f"Checking directories:")
directories = [shuffled_images_path, shuffled_labels_path, train_images_path, val_images_path, train_labels_path, val_labels_path]
for directory in directories:
    if os.path.exists(directory):
        print(f"Directory exists: {directory}")
    else:
        print(f"Directory does NOT exist: {directory}")

# Create directories if they don't exist
os.makedirs(train_images_path, exist_ok=True)
os.makedirs(val_images_path, exist_ok=True)
os.makedirs(train_labels_path, exist_ok=True)
os.makedirs(val_labels_path, exist_ok=True)

# Get all image files and shuffle them
image_files = [f for f in os.listdir(shuffled_images_path) if os.path.isfile(os.path.join(shuffled_images_path, f)) and f.endswith('.png')]
print(f"Image files found: {len(image_files)}")
print(f"First 5 image files: {image_files[:5]}")
random.shuffle(image_files)

# Split the image files into train and val sets
train_count = floor(0.7 * len(image_files))
train_files = image_files[:train_count]
val_files = image_files[train_count:]
print(f"Train files count: {len(train_files)}")
print(f"Validation files count: {len(val_files)}")

# Function to check and copy files
def copy_files(files, src_image_dir, src_label_dir, dst_image_dir, dst_label_dir):
    for file in files:
        image_src_path = os.path.join(src_image_dir, file)
        image_dst_path = os.path.join(dst_image_dir, file)
        label_file = file.replace('.png', '.txt')
        label_src_path = os.path.join(src_label_dir, label_file)
        label_dst_path = os.path.join(dst_label_dir, label_file)
        
        print(f"Copying image: {image_src_path} -> {image_dst_path}")
        if os.path.exists(image_src_path):
            shutil.copy(image_src_path, image_dst_path)
        else:
            print(f"Image file does not exist: {image_src_path}")
        
        print(f"Checking label: {label_src_path}")
        if os.path.exists(label_src_path):
            print(f"Copying label: {label_src_path} -> {label_dst_path}")
            shutil.copy(label_src_path, label_dst_path)
        else:
            print(f"Label file does not exist: {label_src_path}")

# copy train files
copy_files(train_files, shuffled_images_path, shuffled_labels_path, train_images_path, train_labels_path)

# copy val files
copy_files(val_files, shuffled_images_path, shuffled_labels_path, val_images_path, val_labels_path)

print("Splitting completed!")
```

```python
import cv2
import os

# Directory containing the frames and the name of the output video file
frames_dir = '/home/trainee/Desktop/Final_Work/yolov5-master/runs/detect/exp10'  # Update with your frames directory
output_video = 'YOLOv5s-1.avi'        # Update with your desired output video file name

# Get the list of frames and sort them
frames = [os.path.join(frames_dir, f) for f in os.listdir(frames_dir) if f.endswith('.png')]  # Assuming frames are in PNG format
frames.sort()  # Sorting the frames to ensure the correct order

# Read the first frame to get the frame size
frame = cv2.imread(frames[0])
height, width, layers = frame.shape
size = (width, height)

# Create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for .mp4
out = cv2.VideoWriter(output_video, fourcc, 25.0, size)  # 30.0 is the frame rate

# Write each frame to the video
for frame_path in frames:
    frame = cv2.imread(frame_path)
    out.write(frame)

# Release the VideoWriter object
out.release()
print("Video created successfully!")
```
