```python
import cv2
import os

# Directory containing the frames and the name of the output video file
frames_dir = '/home/trainee/Desktop/Final_Work/yolov5-master/runs/detect/exp3'  # Update with your frames directory
output_video = 'output_video_YOLOv5l.avi'        # Update with your desired output video file name

# Get the list of frames and sort them
frames = [os.path.join(frames_dir, f) for f in os.listdir(frames_dir) if f.endswith('.png')]  # Assuming frames are in PNG format
frames.sort()  # Sorting the frames to ensure the correct order

# Read the first frame to get the frame size
frame = cv2.imread(frames[0])
height, width, layers = frame.shape
size = (width, height)

# Create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for .mp4
out = cv2.VideoWriter(output_video, fourcc, 25.0, size)  # 30.0 is the frame rate

# Write each frame to the video
for frame_path in frames:
    frame = cv2.imread(frame_path)
    out.write(frame)

# Release the VideoWriter object
out.release()
print("Video created successfully!")
```
